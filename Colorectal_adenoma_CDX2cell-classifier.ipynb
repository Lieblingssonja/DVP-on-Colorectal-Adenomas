```python
### Classifier based on 6 conditions

def classify_adv(rg_z, rgb_ratio, cellsize, sum_intensities, blue, ellips,
                 
                 #Sum of red and green. Outlier were removed. Values were z_scored.
                 
                 rg_zscored=[-2, -1, 0, 1, 2], 
                 
                 #Ratio_threshold describes the normalized part of blue within a cell. When the ratio is high, the relative amount of
                 #blue within a cell is small. The cells appear brown - epithelial cells.
                 
                 ratio_threshold=[C1_cut['ratio_GR/B'].quantile(3/4), C1_cut['ratio_GR/B'].quantile(0.5)],
                 
                 #Nucleus volumenatry. Here, artifacts are removed.
                 
                 cell_size=[15, 20, 25, 30, 50],
                 
                 #Sum of all RGB color intensities, mean. Filtering 5% of a distribution will remove very dark cells
                 #or artifacts. This will be applied as condition for class 1 cells only.
                 
                 sum_rgb=[C1_cut['sum_intensity-mean'].quantile(0.05), 0],
                 
                 #The mean blue intensity. Only 25% percent will be taken, as those are show the strongest blue.
                 #This should help to distinguish between very blue epithelial cells and ECM cells.
                 
                 blue_int=[C1_cut['NUCLEUS INTENSITY-MEAN TL Brightfield: Blue'].quantile(1/4), C1_cut['NUCLEUS INTENSITY-MEAN TL Brightfield: Blue'].quantile(3/4)],
                 
                 elipsmax= [C1_cut['NUCLEUS ELLIPSE-MAJOR AXIS LENGTH'].median(), 36]):
    
    if rg_z <= rg_zscored[1] and rg_z >= rg_zscored[0] and rgb_ratio >= ratio_threshold[0] and cellsize >= cell_size[2] and sum_intensities >= sum_rgb[0]:
    #and blue >= blue_int[1]:, will turn red
        return 1
    
    if rg_z >= rg_zscored[1] and rg_z <= rg_zscored[3] and rgb_ratio >= ratio_threshold[0] and cellsize >= cell_size[2] and sum_intensities >= sum_rgb[1]:
    #and blue >= blue_int[1]: will turn orange
        return 2
    
    if rg_z >= rg_zscored[3] and rg_z <= rg_zscored[4] and rgb_ratio >= ratio_threshold[1] and cellsize >= cell_size[2]:
    #and sum_intensities >= sum_rgb[1]: will turn yellow
        return 3
    
    if rg_z <= rg_zscored[4] and rgb_ratio <= ratio_threshold[1] and cellsize >= cell_size[1] and blue >= blue_int[0] and ellips <= elipsmax[1]:
        # very blue, rg negative ECM cells
        return 0
    return 5

```

```python
classified_C1 = pd.DataFrame(C1_2)

rg_z_list = list(classified_C1['RGsum_zscore'])
rgb_ratio_list = list(classified_C1['ratio_GR/B'])
cellsize_list = list(classified_C1['NUCLEUS VOLUMETRY'])
rgb_list = list(classified_C1['sum_intensity-mean'])
blue_list = list(classified_C1['NUCLEUS INTENSITY-MEAN TL Brightfield: Blue'])
ellips_list = list(classified_C1['NUCLEUS ELLIPSE-MAJOR AXIS LENGTH'])
```

```python
classificationList = []
df_number = len(rgb_ratio_list)
# Go through all rows and classify them.
for i in range(0, df_number):
    # Save classification in list
    classificationList.append(classify_adv(rg_z_list[i], rgb_ratio_list[i],
                                           cellsize_list[i], rgb_list[i], blue_list[i],
                                          ellips_list[i]))

# Add classification in dataframe
classified_C1["class_101"] = classificationList
classified_C1
```
